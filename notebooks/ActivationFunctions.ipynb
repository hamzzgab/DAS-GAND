```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def linear(x):
    return x

def binary(x):
    return np.where(x < 0, 0, 1)

def sigmoid(x):
    return (1 / (1 + pow(np.e, -x)))

def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def softplus(x):
    return np.log(1 + np.exp(x))

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e_x / np.sum(e_x, axis=-1, keepdims=True)

def relu(x):
    return np.where(x < 0, 0, x)

def lrelu(x, alpha=0.01):
    return np.where(x < 0, alpha * x, x)
```

```python
X = np.linspace(start=-10, stop=10, num=100)

FONT_SIZE = 16

sns.set_theme(style="ticks", palette="flare")
fig, axes = plt.subplots(4,2, sharex=True, sharey=False, figsize=(10,10))
fig.suptitle('Machine Learning Activation Functions', fontsize=FONT_SIZE+3, fontweight='bold')

y = linear(X)
axes[0, 0].plot(X, y)
axes[0, 0].get_xaxis().set_visible(False)
axes[0, 0].set_title('Linear', fontsize=FONT_SIZE)
axes[0, 0].set_yticks(np.linspace(min(y), max(y), 5))

# y = binary(X)
# axes[0, 1].plot(X, y)
# axes[0, 1].get_xaxis().set_visible(False)
# axes[0, 1].set_title('Binary Step', fontsize=FONT_SIZE)
# axes[0, 1].set_yticks(np.linspace(min(y), max(y), 5))


# y = relu(X)
# axes[1, 0].plot(X, y)
# axes[1, 0].get_xaxis().set_visible(False)
# axes[1, 0].set_title('ReLU', fontsize=FONT_SIZE)
# axes[1, 0].set_yticks(np.linspace(min(y), max(y), 5))
# axes[1, 0].annotate('(0, 0)', xy=(0, 0))

# alpha=0.2
# y = lrelu(X, alpha=0.2)
# axes[1, 1].plot(X, y)
# axes[1, 1].get_xaxis().set_visible(False)
# axes[1, 1].set_title('Leaky ReLU', fontsize=FONT_SIZE)
# axes[1, 1].set_yticks(np.linspace(min(y), max(y), 5))
# axes[1, 1].annotate(f'({alpha}, 0)', xy=(alpha, 0))

# y = sigmoid(X)
# axes[2, 0].plot(X, y)
# axes[2, 0].get_xaxis().set_visible(False)
# axes[2, 0].set_title('Sigmoid', fontsize=FONT_SIZE)
# axes[2, 0].set_yticks(np.linspace(min(y), max(y), 5))

y = tanh(X)
axes[2, 1].plot(X, y)
axes[2, 1].get_xaxis().set_visible(False)
axes[2, 1].set_title('Tanh', fontsize=FONT_SIZE)
axes[2, 1].set_yticks(np.linspace(min(y), max(y), 5))

# y = softmax(X)
# axes[3, 0].plot(X, y)
# axes[3, 0].set_title('Softmax', fontsize=FONT_SIZE)
# axes[3, 0].set_yticks(np.linspace(min(y), max(y), 5))

# y = softplus(X)
# axes[3, 1].plot(X, y)
# axes[3, 1].set_title('Softplus', fontsize=FONT_SIZE)
# axes[3, 1].set_yticks(np.linspace(min(y), max(y), 5))

plt.tight_layout()
plt.show();
```

```python
X = np.linspace(start=-10, stop=10, num=100)

FONT_SIZE = 30
LINEWIDTH = 3
plt.rcParams.update({'font.size': FONT_SIZE})

sns.set_theme(style="whitegrid", palette="flare")
fig, axes = plt.subplots(1, 3, sharex=False, sharey=False, 
                         figsize=(20,6))
# fig.suptitle('Machine Learning Activation Functions', fontsize=FONT_SIZE+3, fontweight='bold')

axes.flatten()
y = linear(X)
axes[0].plot(X, y, linewidth=LINEWIDTH)
# axes[0, 0].get_xaxis().set_visible(False)
axes[0].set_title('Linear', fontsize=FONT_SIZE)
axes[0].set_yticks(np.linspace(min(y), max(y), 5))
axes[0].set_xticks(np.linspace(min(X), max(X), 5))
axes[0].tick_params(axis='both', which='major', labelsize=FONT_SIZE-10)
axes[0].tick_params(axis='both', which='minor', labelsize=FONT_SIZE-10)

y = relu(X)
axes[1].plot(X, y, linewidth=LINEWIDTH)
# axes[1, 0].get_xaxis().set_visible(False)
axes[1].set_title('ReLU', fontsize=FONT_SIZE)
axes[1].set_yticks(np.linspace(min(y), max(y), 5))
axes[1].set_xticks(np.linspace(min(X), max(X), 5))
axes[1].annotate('(0, 0)', xy=(0, 0), fontsize=FONT_SIZE)
axes[1].tick_params(axis='both', which='major', labelsize=FONT_SIZE-10)
axes[1].tick_params(axis='both', which='minor', labelsize=FONT_SIZE-10)

# y = tanh(X)
alpha=0.2
y = lrelu(X, alpha=alpha)
axes[2].plot(X, y, linewidth=LINEWIDTH)
# axes[2, 0].get_xaxis().set_visible(False)
axes[2].set_title('Leaky ReLU', fontsize=FONT_SIZE)
axes[2].set_yticks(np.linspace(min(y), max(y), 5))
axes[2].set_xticks(np.linspace(min(X), max(X), 5))
axes[2].annotate(f'({alpha}, 0)', xy=(alpha, 0), fontsize=FONT_SIZE)
axes[2].tick_params(axis='both', which='major', labelsize=FONT_SIZE-10)
axes[2].tick_params(axis='both', which='minor', labelsize=FONT_SIZE-10)


plt.tight_layout()
plt.savefig('ReLU-LeakyRelu-Activation.png', bbox_inches='tight')
plt.show();
```
